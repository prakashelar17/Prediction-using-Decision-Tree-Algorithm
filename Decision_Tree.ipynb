```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
#laod dataset 
path = r'E:\Spark_found\Iris.csv'
data = pd.read_csv(path)
```

```python
#Encoding target data
labelencoder = LabelEncoder()
data['Species'] = labelencoder.fit_transform(data['Species'])
```

```python
#split data into feature and output
X = np.array(data.iloc[: ,1:5])
Y = data['Species']

```

```python
#importing Decision Tree model
model = DecisionTreeClassifier()
```

```python
#spliting data training and test data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
#training model
result = model.fit(X_train,Y_train)
```

```python
# see split of decision tree 
fig = plt.gcf()
fig.set_size_inches(15,8)
tree.plot_tree(result , filled = True);
```

```python
#predict output using test data
Y_pred = model.predict(X_test)
print(Y_pred)
```

```python
acc=accuracy_score(Y_test,Y_pred)
```

```python
print("accuracy of model is :", acc)
```

```python
# prediction for any single test data
Y_pred = model.predict(X_test[0].reshape(1,4))

d ={ 0:'Iris-setosa', 1:'Iris-versicolor',2:'Iris-virginica'}
print('Species of given plant :', d[int(Y_pred)])
```
